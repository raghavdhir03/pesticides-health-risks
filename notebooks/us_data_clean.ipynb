Load in Data

```python
from pathlib import Path
import os

parquet_dir = Path("../data/parquet")

for f in parquet_dir.glob("pdp_*_merged.parquet"):
    # Extract the year from the filename
    year = f.stem.split("_")[1]   # "1992" from "pdp_1992_merged"
    new_name = parquet_dir / f"{year}.parquet"
    os.rename(f, new_name)
    print(f"Renamed {f.name} -> {new_name.name}")
```

```python
pd.set_option('display.max_columns', None)
df = pd.read_csv('/Users/rdhir/Documents/pesticides-health-risks/data/USPesticides_raw.csv')
```

```python
df.head()
```

Split Sample ID into State, Year, Month-Day, Commodity, etc..

```python
df["State"] = df["Sample ID"].str[0:2]          # 2-letter state
df["Year"] = df["Sample ID"].str[2:4]           # 2-digit year
df["Month"] = df["Sample ID"].str[4:6]          # 2-digit month
df["Day"] = df["Sample ID"].str[6:8]            # 2-digit day
df["Commodity_Code"] = df["Sample ID"].str[12:14]  # 2-letter commodity code
df["Lab_Code"] = df["Sample ID"].str[14:17]     # 3-char lab code

# (Optional) combine Month-Day into a single column if you want
df["Month-Day"] = df["Month"] + "-" + df["Day"]
```
